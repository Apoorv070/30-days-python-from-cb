## Multivariate Normal Distribution

```python
##Bivararte
##we need mean vector and covarience vector
```

```python
mean = np.array([0.0,0.0])
cov = np.array([[1,0],[0,1]])
```

```python
dist = np.random.multivariate_normal(mean,cov,500)##500 points
print(dist)
```

```python
print(dist.shape)
## each point has 2 columns repressenting x and y values 
```

```python
plt.scatter(dist[:,0],dist[:,1])#it is like scatter(x,y) x is first col and y is in 2nd col and rows 
##we have to consider all
##points genrated form 2 normal distribution mean is 0 and it has similar spread in x direction and y direction
#sigmaxx = 1 sigmayy=1 and sigmaxy=0 which means change in x does not affect the y axis
#x and y are independent 
```

```python
#now consider the case of momkey where there are 2 cases x as the height of a monkey and y denote the weight of a monkey 
#now height and weight have a postive corelation bw then means on increase of height weight will also increase in such case x 
#and y are not independent so the sigmaxy is not 0 instead it has a postive corelation so sigmaxy could be 0.5 
```

```python
mean = np.array([0.0,0.0])
cov = np.array([[1,0.5],[0.5,1]])
dist = np.random.multivariate_normal(mean,cov,500)##500 points
plt.scatter(dist[:,0],dist[:,1])#it is like scatter(x,y) x is first col and y is in 2nd col and rows 
```
